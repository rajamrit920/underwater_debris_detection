```python
model_path = '/content/runs/detect/train2/weights/best.pt'  # Update with the correct path

# Load the YOLO model
model = YOLO(model_path)
```

```python
import cv2
from IPython.display import display, Image
import tempfile

# Open the video file using OpenCV
cap = cv2.VideoCapture(video_path)

# Check if video opened successfully
if not cap.isOpened():
    print("Error: Could not open video.")
else:
    frame_count = 0  # Frame counter
    frame_skip = 5   # Process every 5th frame (can adjust this value)
    max_frames = 10  # Limit to process only 10 frames (can adjust this value)

    # Create temporary list to save the annotated frames for later use
    annotated_frames = []

    # Loop through the video frames
    while cap.isOpened() and frame_count < max_frames:
        ret, frame = cap.read()
        if not ret:
            break

        # Skip frames based on frame_skip
        if frame_count % frame_skip == 0:
            # Perform prediction on the current frame
            results = model(frame)

            # Draw bounding boxes on the frame
            annotated_frame = results[0].plot()  # Plot the results (bounding boxes)

            # Convert the annotated frame to jpg and save it in a temporary path
            _, buffer = cv2.imencode('.jpg', annotated_frame)  # Convert frame to jpg
            img_path = tempfile.mktemp(suffix='.jpg')  # Create temporary file
            with open(img_path, 'wb') as f:
                f.write(buffer)

            # Add the annotated frame to the list for later viewing
            annotated_frames.append(img_path)

            # Display the annotated frame
            display(Image(filename=img_path))

        frame_count += 1

    cap.release()

# Optionally, return the list of annotated frames for further processing or saving
print("Annotated frames:", annotated_frames)
```
